```python
total_beta_composition = {"A":0, "R":0, "N":0, "D":0, "C":0, "Q":0, "E":0, "G":0, "H":0, "I":0, "L":0, "K":0, "M":0, "F":0, "P":0, "S":0, "T":0, "W":0, "Y":0, "V":0}
total_alpha_composition = {"A":0, "R":0, "N":0, "D":0, "C":0, "Q":0, "E":0, "G":0, "H":0, "I":0, "L":0, "K":0, "M":0, "F":0, "P":0, "S":0, "T":0, "W":0, "Y":0, "V":0}
total_residue_beta = total_residue_alpha = 0
```

```python
dataset_beta = pd.DataFrame()
dataset_alpha = pd.DataFrame()
header_beta = []
header_alpha = []
for i in range(len(alpha)):
    amino_acids = {"A":0, "R":0, "N":0, "D":0, "C":0, "Q":0, "E":0, "G":0, "H":0, "I":0, "L":0, "K":0, "M":0, "F":0, "P":0, "S":0, "T":0, "W":0, "Y":0, "V":0}
    seq = alpha['seq'][i]
    headera = '|'.join(alpha['header'][i].split('|')[:2])
    header_alpha.append(headera)
    length = len(seq)
    total_residue_alpha += length
    for aa_res in seq:
        total_residue_alpha += 1
        if aa_res in amino_acids.keys():
            amino_acids[aa_res] += 1
            total_alpha_composition[aa_res] += 1
    for aa in amino_acids.keys():
        amino_acids[aa] = amino_acids[aa]/length
    dataset_alpha = dataset_alpha.append(amino_acids, ignore_index = True)
    dataset_alpha['label'] = 'Class A'
    dataset_alpha['header'] = header_alpha

for i in range(len(beta)):
    amino_acids = {"A":0, "R":0, "N":0, "D":0, "C":0, "Q":0, "E":0, "G":0, "H":0, "I":0, "L":0, "K":0, "M":0, "F":0, "P":0, "S":0, "T":0, "W":0, "Y":0, "V":0}
    seq = beta['seq'][i]
    headerb = '|'.join(beta['header'][i].split('|')[:2])
    header_beta.append(headerb)
    length = len(seq)
    total_residue_beta += length
    for aa_res in seq:
        total_residue_beta += 1
        if aa_res in amino_acids.keys():
            amino_acids[aa_res] += 1
            total_beta_composition[aa_res] += 1
    for aa in amino_acids.keys():
        amino_acids[aa] = amino_acids[aa]/length
    dataset_beta = dataset_beta.append(amino_acids, ignore_index = True)
    dataset_beta['label'] = 'Class B'
    dataset_beta['header'] = header_beta
```
